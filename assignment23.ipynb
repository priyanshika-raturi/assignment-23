Question 1: What is Deep Learning? Briefly describe how it evolved and how it differs
from traditional machine learning.

**Deep Learning** is a subset of Machine Learning that uses artificial neural networks with multiple layers (called **deep neural networks**) to automatically learn patterns from large amounts of data. It is inspired by how the human brain processes information.

---

##  Evolution of Deep Learning

1. **Early Neural Networks (1940s–1980s)**

   * First models like the perceptron were introduced.
   * Limited computing power and data → not very successful.

2. **Machine Learning Era (1990s–2000s)**

   * Focus shifted to algorithms like decision trees, SVMs.
   * Required **manual feature extraction**.

3. **Deep Learning Breakthrough (2010s–present)**

   * Availability of big data + powerful GPUs.
   * Models like Artificial Neural Networks with many layers became practical.
   * Major success in image, speech, and language tasks.

---

## Deep Learning vs Traditional Machine Learning

| Feature                | Traditional ML               | Deep Learning               |
| ---------------------- | ---------------------------- | --------------------------- |
| **Feature Extraction** | Manual (human-designed)      | Automatic                   |
| **Data Requirement**   | Works with small/medium data | Needs large data            |
| **Performance**        | Good for simple tasks        | Excellent for complex tasks |
| **Computation**        | Less intensive               | High computational power    |
| **Examples**           | Linear Regression, SVM       | CNN, RNN, Transformers      |



Question 2: Explain the basic architecture and functioning of a Perceptron. What are its
limitations?
##  Perceptron: Architecture & Working

A **Perceptron** is the simplest type of artificial neuron used in Machine Learning and is a basic building block of Artificial Neural Networks.

---

## Basic Architecture

A perceptron consists of:

1. **Input values (x₁, x₂, x₃, …)**
2. **Weights (w₁, w₂, w₃, …)** → importance of each input
3. **Bias (b)** → helps shift the decision boundary
4. **Summation function** → calculates weighted sum
5. **Activation function (step function)** → gives final output

 Mathematical form:

[
y = f\left(\sum w_i x_i + b\right)
]

* If result ≥ 0 → Output = 1
* If result < 0 → Output = 0

---

##  How It Works (Step-by-Step)

1. Inputs are multiplied by their respective weights
2. All values are added together with bias
3. The result passes through a **step activation function**
4. Final output is either **0 or 1** (binary classification)

---

##  Example

* Email spam detection:

  * Inputs: keywords, sender, links
  * Output: Spam (1) or Not Spam (0)



Question 3: Describe the purpose of activation function in neural networks. Compare
Sigmoid, ReLU, and Tanh functions.

##  Purpose of Activation Function

In Artificial Neural Networks, an **activation function** decides whether a neuron should be activated (i.e., produce output) or not.

###  Why it is needed:

*  **Introduces non-linearity** → helps model complex patterns
*  **Controls output range**
*  **Enables deep networks to learn complex relationships**
*  Without it, the network behaves like a simple linear model (no real power)

---

##  Comparison: Sigmoid vs ReLU vs Tanh

### 1️⃣ Sigmoid Function

[
\sigma(x) = \frac{1}{1 + e^{-x}}
]

* Output range: **0 to 1**
* Shape: S-curve

**✔️ Advantages:**

* Good for probability outputs (e.g., classification)

**❌ Disadvantages:**

* Vanishing gradient problem
* Slow learning for deep networks

---

### 2️⃣ Tanh Function

[
\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
]

* Output range: **-1 to 1**
* Zero-centered

**✔️ Advantages:**

* Better than sigmoid (centered around 0)
* Stronger gradients

**❌ Disadvantages:**

* Still suffers from vanishing gradient

---

### 3️⃣ ReLU (Rectified Linear Unit)

[
f(x) = \max(0, x)
]

* Output range: **0 to ∞**

**✔️ Advantages:**

* Very fast computation
* Solves vanishing gradient (for positive values)
* Most widely used in deep learning

**❌ Disadvantages:**

* “Dying ReLU” problem (neurons can stop learning if output = 0)



Question 4: What is the difference between Loss function and Cost function in neural
networks? Provide examples.

##  Loss Function vs Cost Function

In Artificial Neural Networks, both **loss** and **cost** functions are used to measure how well a model is performing—but they are slightly different in scope.

---

##  Loss Function

* Measures **error for a single training example**
* Tells how far the predicted output is from the actual output

###  Example:

* **Mean Squared Error (MSE) for one data point:**
  [
  L = (y - \hat{y})^2
  ]

 Used during training to evaluate individual predictions

---

## 🔹 Cost Function

* Measures **average loss over the entire dataset**
* It is the value that the model tries to **minimize during training**

###  Example:

* **MSE Cost Function:**
  [
  J = \frac{1}{n} \sum (y - \hat{y})^2
  ]

 Used to update weights using optimization algorithms



Question 5: What is the role of optimizers in neural networks? Compare Gradient
Descent, Adam, and RMSprop.

## Role of Optimizers in Neural Networks

In Artificial Neural Networks, an **optimizer** is an algorithm used to **update the model’s weights and biases** to minimize the **cost function**.

###  Main Role:

* Reduce error (loss/cost)
* Adjust weights in the right direction
* Help the model learn faster and more efficiently

 It works using gradients (from backpropagation) to find the **minimum error point**.

---

##  Common Optimizers Comparison

### 1️⃣ Gradient Descent

* Updates weights using the slope (gradient) of the cost function

[
w = w - \eta \cdot \nabla J
]

(η = learning rate)

**✔️ Advantages:**

* Simple and easy to understand

**❌ Disadvantages:**

* Slow for large datasets
* Can get stuck in local minima
* Same learning rate for all parameters

---

### 2️⃣ RMSprop (Root Mean Square Propagation)

* Adjusts learning rate individually for each parameter
* Uses moving average of squared gradients

**✔️ Advantages:**

* Faster convergence than Gradient Descent
* Works well for non-stationary problems

**❌ Disadvantages:**

* More complex
* Needs tuning

---

### 3️⃣ Adam (Adaptive Moment Estimation)

* Combines **Momentum + RMSprop**
* Uses both:

  * Moving average of gradients
  * Moving average of squared gradients

**✔️ Advantages:**

* Very fast and efficient
* Requires less tuning
* Most widely used optimizer

**❌ Disadvantages:**

* Slightly more memory usage

Question 6: Write a Python program to implement a single-layer perceptron from
scratch using NumPy to solve the logical AND gate.


```python
import numpy as np

# Step 1: Input dataset (AND gate)
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# Step 2: Output labels
y = np.array([0, 0, 0, 1])

# Step 3: Initialize weights and bias
weights = np.zeros(2)
bias = 0
learning_rate = 0.1
epochs = 10

# Step 4: Activation function (Step function)
def step_function(z):
    return 1 if z >= 0 else 0

# Step 5: Training the perceptron
for epoch in range(epochs):
    print(f"Epoch {epoch+1}")
    for i in range(len(X)):
        # Linear combination
        z = np.dot(X[i], weights) + bias

        # Prediction
        y_pred = step_function(z)

        # Error
        error = y[i] - y_pred

        # Update weights and bias
        weights += learning_rate * error * X[i]
        bias += learning_rate * error

        print(f"Input: {X[i]}, Predicted: {y_pred}, Actual: {y[i]}")

# Step 6: Final weights and bias
print("\nFinal Weights:", weights)
print("Final Bias:", bias)

# Step 7: Testing
print("\nTesting:")
for i in range(len(X)):
    z = np.dot(X[i], weights) + bias
    y_pred = step_function(z)
    print(f"Input: {X[i]} => Output: {y_pred}")
```

Question 7: Implement and visualize Sigmoid, ReLU, and Tanh activation functions
using Matplotlib.
(Include your Python code and output in the code box below.)

```python
import numpy as np
import matplotlib.pyplot as plt

# Create input values
x = np.linspace(-10, 10, 100)

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Compute outputs
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)

# Plot the functions
plt.figure(figsize=(10, 6))

plt.plot(x, y_sigmoid, label='Sigmoid', linestyle='-')
plt.plot(x, y_relu, label='ReLU', linestyle='--')
plt.plot(x, y_tanh, label='Tanh', linestyle='-.')

# Add labels and legend
plt.title("Activation Functions Visualization")
plt.xlabel("Input (x)")
plt.ylabel("Output")
plt.legend()
plt.grid()

# Show plot
plt.show()
```

Question 8: Use Keras to build and train a simple multilayer neural network on the
MNIST digits dataset. Print the training accuracy.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Normalize data (0–255 → 0–1)
x_train = x_train / 255.0
x_test = x_test / 255.0

# Flatten images (28x28 → 784)
x_train = x_train.reshape(-1, 28 * 28)
x_test = x_test.reshape(-1, 28 * 28)

# Build model
model = keras.Sequential([
    layers.Dense(128, activation='relu', input_shape=(784,)),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Compile model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train model
history = model.fit(x_train, y_train, epochs=5, batch_size=32)

# Print training accuracy (last epoch)
train_accuracy = history.history['accuracy'][-1]
print("Training Accuracy:", train_accuracy)

```

Question 9: Visualize the loss and accuracy curves for a neural network model trained
on the Fashion MNIST dataset. Interpret the training behavior.


```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt

# Load Fashion MNIST dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()

# Normalize data
x_train = x_train / 255.0
x_test = x_test / 255.0

# Build model
model = keras.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Compile model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Train model
history = model.fit(
    x_train, y_train,
    epochs=10,
    batch_size=32,
    validation_data=(x_test, y_test)
)

# Plot Accuracy
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy Curve')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid()

# Plot Loss
plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Curve')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid()

plt.show()
```

#  1. Model Design (Perceptron vs Multilayer NN)

Since fraud detection is **complex and non-linear**, a **Multilayer Neural Network (MLP)** is better than a single perceptron.

### 🔹 Model Architecture:

* **Input Layer**: All features (amount, merchant ID, location, etc.)
* **Hidden Layers**:

  * Dense (128 neurons, ReLU)
  * Dense (64 neurons, ReLU)
  * Dense (32 neurons, ReLU)
* **Output Layer**:

  * 1 neuron (Sigmoid) → Binary classification (Fraud / Legit)

 This allows the model to capture **complex patterns in transactions**

---

#  2. Activation & Loss Function

### 🔹 Activation Function:

* **ReLU (Hidden Layers)**
  ✔ Fast
  ✔ Avoids vanishing gradient
  ✔ Works well for deep models

* **Sigmoid (Output Layer)**
  ✔ Outputs probability (0–1)
  ✔ Suitable for binary classification

---

### 🔹 Loss Function:

* **Binary Cross-Entropy**
  [
  Loss = -[y \log(p) + (1-y)\log(1-p)]
  ]

✔ Best for binary classification
✔ Penalizes wrong predictions strongly

 For imbalance:

* Use **Weighted Binary Cross-Entropy** (higher weight for fraud class)

---

#  3. Training & Evaluation (Handling Imbalance)

Fraud datasets are **highly imbalanced** (e.g., 99% legit, 1% fraud)

### 🔹 Techniques to Handle Imbalance:

* **Class Weights**
  → Give more importance to fraud class

* **Oversampling** (e.g., SMOTE)

* **Undersampling** (reduce majority class)

---

### 🔹 Evaluation Metrics (IMPORTANT ❗)

Accuracy is misleading

Use:

* **Precision** → How many predicted frauds are correct
* **Recall** → How many actual frauds detected
* **F1-score** → Balance of precision & recall
* **ROC-AUC Score**

 In banking, **Recall is most important** (catch maximum fraud)

---

#  4. Optimizer & Overfitting Prevention

### 🔹 Optimizer:

* **Adam (Adaptive Moment Estimation)**
  ✔ Fast convergence
  ✔ Handles large datasets well
  ✔ Automatically adjusts learning rate

---

### 🔹 Prevent Overfitting:

* **Dropout Layers** (e.g., 0.3–0.5)
* **Early Stopping** (stop when validation loss increases)
* **Regularization (L2)**
* **Cross-validation**
* **Feature selection**
